city와 f4를 기준으로 f5의 평균값을 구한 다음, f5를 기준으로 상위 7개 값을 모두 더해 출력하시오 (소수점 둘째자리까지 출력)

```python
import pandas as pd

df = pd.read_csv('basic1.csv')

```

```python
df.groupby(['city', 'f4'])[['f5']].mean().sort_values(by = 'f5' , ascending = False)[:7].sum()
```

```python
df = df.groupby(['city', 'f4'])[['f5']].mean()
```
