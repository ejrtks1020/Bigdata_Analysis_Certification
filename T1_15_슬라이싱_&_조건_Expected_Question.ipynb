```python
# 주어진 데이터 셋에서 age컬럼 상위 20개의 데이터를 구한 다음 
# f1의 결측치를 중앙값으로 채운다.
# 그리고 f4가 ISFJ와 f5가 20 이상인 
# f1의 평균값을 출력하시오!
```

```python
import pandas as pd

df = pd.read_csv('basic1.csv')
```

```python
df = df.sort_values('age', ascending = False).head(20)
```

```python
df['f1'].fillna(df['f1'].median(), inplace = True)
```

```python
df[(df['f4'] == 'ISFJ') & (df['f5'] >= 20)]['f1'].mean()
```
